```python
import csv
from openpyxl import Workbook

class CSVToExcelConverter:
    def __init__(self, csv_file_path, excel_file_path):
        self.csv_file_path = csv_file_path
        self.excel_file_path = excel_file_path
        self.wb = Workbook()
        self.wb.remove(self.wb.active)  # 移除默认sheet

    def read_csv(self):
        with open(self.csv_file_path, mode='r', encoding='utf-8-sig') as file:
            reader = csv.DictReader(file)
            self.data = [row for row in reader]

    def write_to_excel(self):
        years = sorted(
            set(row['年份'] for row in self.data),
            key=lambda x: int(x.replace('年', ''))
        )
        for year in years:
            self.create_sheet_for_year(year)

    def create_sheet_for_year(self, year):
        sheet = self.wb.create_sheet(title=year)
        headers = self.data[0].keys()
        for col_idx, header in enumerate(headers, start=1):
            sheet.cell(row=1, column=col_idx).value = header
        row_idx = 2  
        for row in self.data:
            if row['年份'] == year:
                if any(row[header] for header in headers):  # 检查行是否为空
                    for col_idx, header in enumerate(headers, start=1):
                        sheet.cell(row=row_idx, column=col_idx).value = row[header]
                    row_idx += 1

    def save_excel(self):
        self.wb.save(self.excel_file_path)

# 使用示例
csv_file_path = r'/home/user2/pythontext/2012_2021.csv'
excel_file_path = r'/home/user2/pythontext/res.xlsx'

converter = CSVToExcelConverter(csv_file_path, excel_file_path)
converter.read_csv()
converter.write_to_excel()
converter.save_excel()
```
